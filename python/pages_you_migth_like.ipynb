```python
import json #importing json to working with json data
```

```python
def load_json_data(filename):
    with open(filename,"r") as f:
        return json.load(f)
```

```python
def pages_you_migth_like(user_id,data):
    users_pages = {} #stroing users and their liked pages

    for user in data["users"]:
        users_pages[user["id"]] = set(user["liked_pages"])

    if user_id not in users_pages:
        return []
        
    given_user_liked_pages = users_pages[user_id] 

    page_suggestions = {}

    for other_user , pages in users_pages.items():
        
        if other_user != user_id :
            
            shared_pages = given_user_liked_pages.intersection(pages) #taking the common liked pages between given user and other user
            
            for page in pages:
                if page not in given_user_liked_pages:
                    page_suggestions[page] = page_suggestions.get(page,0) + len(shared_pages)
                    
    sortedd_page_suggestions = sorted(page_suggestions.items() , key= lambda x:x[1] , reverse = True)
    
    return [page_id for page_id ,_ in sortedd_page_suggestions ]
                    
            
    
```

```python
data = load_json_data("massive_json_data.json")
user_id = 11
reccomendation = pages_you_migth_like(user_id,data)
print(f"page recommendation for {user_id} is : {reccomendation}")
```
