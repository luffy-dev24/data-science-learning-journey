```python
import json

def load_json_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
#finding the people with max posts

max_posts = 0
person_with_max_post = ""

parsed_data = load_json_data("parsed_data_json.json")

for chunk in parsed_data[1:]:
    
    user_posts = int(chunk["no_of_posts"])
    
    if user_posts > max_posts :
        
        max_posts = user_posts
        person_with_max_post = chunk["username"]
        
print(f"Person with maximum post : {person_with_max_post}")
print(f"Number of  post : {max_posts}")
```

```python
#finding people with maximum follwers
max_followers = 0
person_with_max_followers = ""

parsed_data = load_json_data("parsed_data_json.json")

for chunk in parsed_data[1:]:
    user_followers = int(chunk["no_of_follwers"])

    if user_followers > max_followers:
        max_followers = user_followers
        person_with_max_followers = chunk["username"]
        
print(f"Person with maximum Followers : {person_with_max_followers}")
print(f"Number of  Followers : {max_followers}")
        
```
