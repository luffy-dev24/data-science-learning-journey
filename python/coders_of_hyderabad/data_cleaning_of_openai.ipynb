```python
import json

def load_json_data(filename):
    with open(filename,"r") as f:
        data= f.read()
    return data
```

```python
initial_data = load_json_data("openai_instagram_followers.txt")
chunks = initial_data.split("\n")
```

```python
def parse_chunk(chunk):
    sep_chunk = chunk.split("|")
    username = sep_chunk[0]
    no_of_posts = sep_chunk[1]
    no_of_followers = sep_chunk[2]
    no_of_followings = sep_chunk[3]
    category = sep_chunk[4]
    bio = sep_chunk[5]

    parsed_chunk = {"username":username,"no_of_posts":no_of_posts,"no_of_followers":no_of_followers,
                    "no_of_followings":no_of_followings,"category":category,"bio":bio}
    return parsed_chunk
    
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
```

```python
with open("parsed_data_json.json","w") as f:
    json.dump(all_chunks,f)
```
