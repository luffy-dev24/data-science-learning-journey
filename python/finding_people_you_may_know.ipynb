```python
import json

def load_json_data(filename): #function to load the json data
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
def finding_people_you_may_know(user_id,data):
    
    user_friends_list = {} #dictionary to store the friends of eact user
    
    for user in data["users"]:
        user_friends_list[user["id"]] = set(user["friends"])

    if user_id not in user_friends_list: 
        return [] #returning empty list if given user id is not the user friends list
        
    direct_friends_of_given_user = user_friends_list[user_id]

    suggestions = {}

    for direct_friend in direct_friends_of_given_user:
        
        for mutual_friend in user_friends_list[direct_friend]:
            
            if mutual_friend != user_id and mutual_friend not in direct_friends_of_given_user:
                
                suggestions[mutual_friend] = suggestions.get(mutual_friend,0)+1
    sortedd_suggesstions = sorted(suggestions.items(), key = lambda x:x[1] , reverse = True)
    
    return [userid for userid , _ in sortedd_suggesstions]

                
                
        
```
