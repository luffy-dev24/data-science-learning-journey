```python
import json #importing the json module to work with JSON data
```

```python
#json.dump function write the json on a file 
```

```python
with open("data.json","w") as f:
    json.dump(data,f)
```

```python
#creating a function to load the json data
```

```python
def load_json_data(filename):
    #loading the json data from a file using json.load
    with open(filename,"r") as f:
        data = json.load(f)
        return data
```

```python
loaded_data = load_json_data("data.json")
```
