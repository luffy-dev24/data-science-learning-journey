```python
def load_json_data(filename):  #function to load the json data from a file
    with open(filename,"r") as f:
        data = json.load(f)
        return data
```

```python
def get_friends_name(names_list,data): #function to get names of friends
    friends_list = []
    for user in data["users"]:
        if user["id"] in names_list:
            friends_list.append(user["name"])
    return friends_list
    
```

```python
def get_pages_names(pages_list_id , data): #function to get pages names
    pages_list = [] 
    for page in data["pages"]:
        if page["id"] in pages_list_id:
            pages_list.append(page["name"])
    return pages_list
    
```

```python
def display_user_details() : # function to display the user details
    data = load_json_data("data.json")
    print("Users Information : " )
    print()
    for user in data["users"]:
        print(f"User Id : {user["id"]} , Username : {user["name"]} , Friends : { get_friends_name(user["friends"],data)} Liked Pages : {get_pages_names(user["liked_pages"],data)} ")
        print()
    print("-"*100)
    print("Pages Information : ")
    print()
    for page in data["pages"]:
        print(f"Page Id : {page["id"]} , Page Name : {page["name"] } ")
        print()
```
